```python
import numpy as np
import matplotlib.pyplot as plt

# Our data
# X represents the features (pizza size)
X = np.array([6, 8, 10, 14, 18])
# y represents the labels (pizza price)
y = np.array([7, 9, 13, 17.5, 18])

# Let's visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', marker='x')
plt.title('Pizza Price vs. Size')
plt.xlabel('Size (inches)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

**Step 2: The Model's "Engine" (Defining our Functions)**

Now, let's translate the math we discussed into Python functions.

**A. The Hypothesis (Our Prediction Function)**
This function simply takes our features (X) and our parameters (
theta_0,
theta_1) and calculates the predicted price according to the line's equation.

**B. The Cost Function (Our "Badness Score")**
This function implements the Mean Squared Error (MSE) formula. It tells us how wrong our model is for a given set of parameters.

**C. The Gradient Descent "Nudger"**
This is the main learning function. It takes our data and iteratively "nudges" our parameters (
theta_0,
theta_1) to minimize the cost. I've included the calculus part (the derivatives `d_theta0` and `d_theta1`) for youâ€”you can treat them as the "downhill direction" calculator we discussed.

```python
# Initialize parameters (start with a random guess)
# theta_0 is our "base price" guess
# theta_1 is our "price per inch" guess
theta_0 = 0.0
theta_1 = 0.0

# Define our learning rate and number of iterations
learning_rate = 0.001
iterations = 1001

# Get the number of training examples
m = len(X)

# --- The Learning Loop (Gradient Descent) ---
cost_history = [] # To store the cost at each iteration

for i in range(iterations):
    # 1. Make a prediction using the current parameters
    # This is our Hypothesis: h(x) = theta_0 + theta_1 * x
    predictions = theta_0 + theta_1 * X

    # 2. Calculate the error
    error = predictions - y

    # 3. Calculate the cost (our "badness score") for this iteration
    # This is our Cost Function: J = (1/2m) * sum(error^2)
    cost = np.sum(error**2) / (2 * m)
    cost_history.append(cost)

    # 4. Calculate the "downhill direction" (the gradients)
    d_theta0 = (1/m) * np.sum(error)
    d_theta1 = (1/m) * np.sum(error * X)

    # 5. Update the parameters (take a step downhill)
    # This is the "nudging" part
    theta_0 = theta_0 - learning_rate * d_theta0
    theta_1 = theta_1 - learning_rate * d_theta1

    # Print the cost every 100 iterations to see the progress
    if (i % 100) == 0:
        print(f"Iteration {i}: Cost = {cost:.2f}, theta_0 = {theta_0:.2f}, theta_1 = {theta_1:.2f}")

print("\n--- Training Complete ---")
print(f"Final Learned Parameters:")
print(f"theta_0 (Base Price) = {theta_0:.2f}")
print(f"theta_1 (Price per Inch) = {theta_1:.2f}")
```

```python
# Plot the original data
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', marker='x', label='Actual Prices')

# Plot the best-fit line our model learned
plt.plot(X, theta_0 + theta_1 * X, color='red', label='Predicted Prices (Our Model)')

plt.title('Pizza Price vs. Size - Model Fit')
plt.xlabel('Size (inches)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Predict the price for a 12-inch pizza
new_pizza_size = 12
predicted_price = theta_0 + theta_1 * new_pizza_size

print(f"\nThe predicted price for a {new_pizza_size}-inch pizza is ${predicted_price:.2f}")
```
