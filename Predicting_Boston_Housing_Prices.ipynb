**Step 1: Setup and Load the Data**

* Import the necessary libraries: numpy, matplotlib.pyplot, and sklearn.

* The Boston dataset is available directly within scikit-learn.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LinearRegression
boston = fetch_openml(name="boston", version=1, as_frame=True) # You may need to verify the correct data_id or name/version on OpenML
X = boston.data
y = boston.target
# The dataset is a dictionary-like object. Print its DESCR key to understand what features it contains.
print(boston.DESCR)
```

Select your feature (X). For this project, let's use the feature named 'RM' (average number of rooms per dwelling).

Select your target (y). This is the housing price, which is the target variable.

Crucially: Create a scatter plot of 'RM' vs. price to visually confirm if there's a linear relationshi

```python
X=boston.data['RM']
X=X.values.reshape(-1, 1)
# For reshaping the feature coloumn into the array here -1 used to automatically detect the number of rows while 1 represents 1 column
plt.figure(figsize=(8,6))
plt.scatter(X,y,color='blue',marker='x')
plt.title('Price vs.Number of Rooms ')
plt.xlabel('Avg.no of Rooms')
plt.ylabel('Price($)')
plt.grid(True)
plt.show()
```

* Create an instance of the model (e.g., model = LinearRegression()).

* Train your model using the data you prepared. In scikit-learn, this is done with the .fit() method

```python
# Training the Model
model=LinearRegression().fit(X,y)
print(model.score(X,y))
print(f'Coffecient: {model.coef_[0]:.2f} $')
print(f'Intercept: {model.intercept_:.2f}')
```

**1. The Coefficient / Slope (1)**


**"For each additional room, the model predicts that the median house price increases by 9.1 units."**

Since the dataset description (boston.DESCR) tells us the target value is in '$1000s', we can translate this into a powerful business insight:

**"For each additional room, the price is expected to increase by $9,100."**

**3. The Model's Performance (R Score)**

We calculated a score of 0.4835. This is the R-squared (R) value. In simple terms, this means:

**"The average number of rooms explains about 48% of the variability in Boston housing prices."**


```python
y_predict=model.predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X,y,color='blue',marker='x',label='Actual Data')
plt.plot(X,y_predict,color='red',label='Regression Line')
plt.title('Price vs.Number of Rooms ')
plt.xlabel('Avg.no of Rooms')
plt.ylabel('Price($)')
plt.legend()
plt.grid(True)
plt.show()
```
